##### 2.3.3.1 Columns based on customer behaviour, so need to ignore them:

```python
loan_data.shape
```

#### 2.3.4 Removing Rows
* Removing  "current" from loan status - as these kind of customers are not important for this business case

```python
# those having "loan_status" current are not fully paid nor defaulted , so need to drop them 
loan_data=loan_data[loan_data.loan_status!="Current"]
```

#### 2.3.5 removing special charcters like % for making data type float/int for numeric analysis
* Column int_rate has % sign - so its good to convert it to number by removing % sign

```python
# right now this column was an object due to "%" sign as we need this column as a numeric/float 
# and it is interrest rate on the loan 
# because numeric variables help in calculation analysis without converting to numberic/float cant perform any mathmatical operation
loan_data['int_rate'] = loan_data['int_rate'].str.strip('%').astype(float)
# right now this column was an object due to "%" sign as we need this column as a numeric/float 
# and it is Revolving line utilization rate, or the amount of credit the borrower is using relative to all available revolving credit.
# because numeric variables help in calculation analysis without converting to numberic/float cant perform any mathmatical operation
loan_data['revol_util'] = loan_data['revol_util'].str.strip('%').astype(float)
```

#### 2.3.6 removing special charcters like + and < for making data type float/int for numeric analysis
* Column `emp_length` has special character - thus it needs to be cleaned with just int values (and consider <1 as 0)

```python
# we need may perform binning for data analysis, so we have replace texual/special characters to make it interger from object 
loan_data['emp_length'] = loan_data['emp_length'].str.strip(' years')
# possible values are lies within 0 to 10 to and so 10+ will be consider as 10 , so removing "+" sign
loan_data['emp_length'] = loan_data['emp_length'].str.strip('+')
# length will not be in -ve so less then 1 simply indicated lenngth is zero , so better to replace "<1" with zero 
loan_data['emp_length'] = loan_data['emp_length'].replace('< 1','0')
```

#### 2.3.7 Converting column to date format and deriving year and month from it
* Convert `issue_d` column associated with dates to date format and extract derive columns from those like `year`, `month` etc

```python
# date will play imprtant role to predict time series graph and in filtering data by date , year , month etc.
# so perform all actions related to dat in data analysis first fixing the format of date in "issue_d" column
# this column is help in data analysis to checking date wise data 
loan_data['issue_d'] = pd.to_datetime(loan_data['issue_d'],format='%b-%y') 
# to_datetime will convert to porper format "year-month-date"
```

```python
# creating a new column which contain issue year with the help of DatetimeIndex method
loan_data['issue_year'] = pd.DatetimeIndex(loan_data['issue_d']).year
# creating a new column which contain issue month with the help of DatetimeIndex method
loan_data['issue_month'] = pd.DatetimeIndex(loan_data['issue_d']).month
```

#### 2.3.8 Merging two similar values in a column
* `home_ownership` Convert `other` and None as `Rent` Home ownwer

```python
# those "home_ownership" is NONE is  Convert other and None as Rent Home owner
# reason to change this NONE values are 3 very less and also if the owner not own the home then high chance to be in a 
# RENT
loan_data['home_ownership'] = ["RENT" if i=='NONE' else i for i in loan_data['home_ownership']]
```

#### 2.3.9 Inputing values from columns where there are few blanks/nulls
* Remove nulls from remaining columns (if any)

```python
# cheking null values count in column , so heavy null values need to be fixed
loan_data.isnull().sum().sort_values(ascending=False)
```

* then null values columns need to be fixed 

```python
# checking how is the distribution of pub_rec_bankruptcies so we can choose right metric for impute
sns.histplot(data=loan_data['pub_rec_bankruptcies'])
plt.show()
```

```python
# 2.3.9.1 For imputing 'pub_rec_bankruptcies' as most of the people have no bankruptcies - so its good to use mode in this case
#imputing pub_rec_bankruptcies for nulls with mode price
loan_data['pub_rec_bankruptcies'].fillna(loan_data['pub_rec_bankruptcies'].mode()[0],inplace=True)
```

```python
# checking how is the distribution of emp_length so we can choose right metric for impute
sns.histplot(data=loan_data['emp_length'])
plt.show()
```

For imputing `emp_length` we can not take 10 years - as we have more than 1000 rows with nulls. This might deviate the mean and other metrics for whole data. So its better to take median, which will take few population from 10 years, few from 1-3 years of emp experience


```python
# 2.3.9.2 imputing median for nulls with median price
loan_data['emp_length'].fillna(round(loan_data['emp_length'].median()),inplace=True)
loan_data['emp_length'] = loan_data['emp_length'].astype(int)
```

```python
# 2.3.9.3 fiiling null values with most occuring values in revol_util
# as revol_util need to fill null values mean and median is not taken 
# because this is we consider as categorial column 
loan_data['revol_util'].fillna(loan_data['revol_util'].mode()[0],inplace=True)
```

```python
# checking again all column null values sum count
loan_data.isnull().sum().sort_values(ascending=False)
```

# Final insights after Analysis:

* As we have noticed that maximum defaults are from loans which are provided int the month of `May`, `Sep`, `Oct`, `Dec`, thus Bank has to be careful with loan application, in these months
* `small_business` has highest percentage of default and `debt consolidation` has maximun number of defaults, So these two need to be check carefully
*  As we have noticed that maximum defaults are from loans where there was either high loan granted or loan with very high interest rates thus High `loan amount` and High `interest Rate` should be avoided to be given to avoid defaulters
* Bank Need to more carefull for State `NE` as most defauters are from this state
* for loans with  purpose of `small business` and `debt consolidation` avoid giving loans to `E`, `F` & `G` Grade and give term max 36 months as 60 month term have most defaults in the past
* Bank has to be more careful while granting loans to low salary peoples ranging from `4k to 17.2k` as most defaults has been seen form these categories
